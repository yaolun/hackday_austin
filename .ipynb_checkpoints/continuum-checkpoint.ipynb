```python
%pylab inline
def cont_mask(wl, flux, sigma=None, numpix=20, maxiter=100):
    """
    Need: numpy, astropy.
    Return a tuple with continuum and the mask which label all of the bad pixel (NaN) and statistical outlier as False
    """
    import numpy as np
    import matplotlib.pyplot as plt
    from djs_reject import djs_reject
    from astropy.convolution import convolve, Box1DKernel
    from scipy.interpolate import interp1d
    
    if sigma == None:
        [upper, lower] = [3.0, 3.0]
    elif len(sigma) == 1:
        [upper, lower] = [float(sigma), float(sigma)]
    else:
        [upper, lower] = [sigma[0], sigma[1]]
    
    xdata = np.array(wl)
    ydata = np.array(flux)
    boxcar_kernel = Box1DKernel(numpix)
    outmask = np.ones_like(ydata)
    ymodel = convolve(ydata, boxcar_kernel)
    
    ystd = np.nanstd(ydata-ymodel)
    ystd = np.full(len(wl), ystd)
    (outmask, qdone)  = djs_reject(ydata, ymodel, outmask=outmask, sigma=ystd, upper=upper, lower=lower)
    
    xgood  = xdata[outmask == True]
    ygood  = ydata[outmask == True]
    err    = ydatastd[outmask == True)

    if (xdata[0] in xgood) == False:
        xgood = np.hstack((xdata[0], xgood))
        ygood = np.hstack((ydata[0], ygood))
    if (xdata[-1] in xgood) == False:
        xgood = np.hstack((xdata[-1], xgood))
        ygood = np.hstack((ydata[-1], ygood))
    ymodel = convolve(ygood, boxcar_kernel)
    f = interp1d(xgood, ymodel, kind='linear')  # kind=cubic
    ymodel = f(xdata)
    
    
    ydata2d  = np.full((len(wl), numpix), np.nan)
    ymodel2d = np.full((len(wl), numpix), np.nan)
    for i in range(0, numpix):
        if i+1 < numpix/2:
            ydata2d[:,i]  = np.hstack((np.full(numpix/2-i-1, np.nan), ydata[0:-(numpix/2-i-1)]))
            ymodel2d[:,i] = np.hstack((np.full(numpix/2-i-1, np.nan), ymodel[0:-(numpix/2-i-1)]))
        elif (numpix-i) <= numpix/2:
            ydata2d[:,i]  = np.hstack((ydata[i-numpix/2+1:], np.full(i-numpix/2+1, np.nan)))
            ymodel2d[:,i] = np.hstack((ymodel[i-numpix/2+1:], np.full(i-numpix/2+1, np.nan)))
        else:
            ydata2d[:,i]  = ydata
            ymodel2d[:,i] = ymodel
    ystd  = np.nanstd(ydata2d-ymodel2d, 1)
    
    print ystd
   
#     ystd  = ystd + ymodel
    ydatastd = np.nanstd(ydata2d, 1)
    
    fig = plt.figure(figsize=(12,8))
    ax = fig.add_subplot(111)
    ax.plot(xdata, ydata-ymodel)
    
#     ax.plot(xdata, ymodel)
#     ax.errorbar(xdata, ydata-ymodel, yerr=ystd)
    ax.set_xlim([60,70])
#     ax.set_ylim([-5,5])
    
    qdone = False
    iiter = 0
    while (qdone == False) and (iiter <= maxiter):
        (outmask, qdone)  = djs_reject(ydata, ymodel, outmask=outmask, sigma=ystd, upper=upper, lower=lower)

        xgood  = xdata[outmask == True]
        ygood  = ydata[outmask == True]
        err    = ydatastd[outmask == True]
        ymodel = convolve(ygood, boxcar_kernel)
        
#         ax.plot(xgood, ygood-iiter*2)
        
        if (xdata[0] in xgood) == False:
            xgood = np.hstack((xdata[0], xgood))
            ygood = np.hstack((ydata[0], ygood))
        if (xdata[-1] in xgood) == False:
            xgood = np.hstack((xdata[-1], xgood))
            ygood = np.hstack((ydata[-1], ygood))
        
        f = interp1d(xgood, ygood, kind='linear')  # kind=cubic
        ymodel = f(xdata)
        
        iiter  = iiter + 1
        if iiter == maxiter:
            print 'fail to converage'
    f = interp1d(ymodel, xdata)
    continuum = convolve(ymodel, boxcar_kernel)
    print iiter
    return (ymodel, outmask)
#     return ymodel


import numpy as np

from astropy.io import ascii
data = ascii.read('/Users/yaolun/bhr71/fitting/0106/pacs/data/BHR71_centralSpaxel_PointSourceCorrected_CorrectedYES_trim.txt')
wl = data['col1']
flux = data['col2']
# cont_mask(wl, flux)
(continuum, outmask) = cont_mask(wl, flux)
```
